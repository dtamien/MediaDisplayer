```python
import requests
import os

def get_data(session, n):
    v = str(n) if n >= 10 else "0" + str(n)
    url = f"https://www.yr.no/assets/images/weather-symbols/light-mode/default/svg/{v}d.svg"
    try:
        r = session.get(url)
        r.raise_for_status()
        with open(f"weather_icons/{v}d.svg", "wb") as f:
            f.write(r.content)
        print(f"Downloaded {v}.svg")
    except Exception as e:
        print(f"Failed to download {v}.svg:", e)

# Create a session with connection pooling
session = requests.Session()

# Create the directory if it doesn't exist
os.makedirs("weather_icons", exist_ok=True)

# Download data using the session
for i in range(1, 1000):
    get_data(session, i)
```
